```python
import shutil

# Directory where your frame files are located
source_directory = '/Data/train/P1'
destination_folder = '/Raw Samples Dataset/Real time mon and QA/Experiment 1'

# Ensure destination folder exists
os.makedirs(destination_folder, exist_ok=True)

# Function to extract frame number from filename
def extract_frame_number(filename):
    try:
        # Extracting frame number from 'P1 (10).png' format
        return int(filename.split('(')[1].split(')')[0])
    except (ValueError, IndexError):
        return float('inf')  # Return infinity if frame number extraction fails

# Get list of files in the directory
files = os.listdir(source_directory)

# Sort files based on extracted frame numbers
sorted_files = sorted(files, key=lambda x: extract_frame_number(x))
total_files = len(sorted_files)
print(total_files)
start_index = (total_files - 5000) // 2
end_index = start_index + 5000

# Copy the selected files to the destination folder
for file_name in sorted_files[start_index:end_index]:
    shutil.copy(os.path.join(source_directory, file_name), os.path.join(destination_folder, file_name))


def copy_middle_jpegs(input_folder, output_folder):
    # Create the output directory if it does not exist
    os.makedirs(output_folder, exist_ok=True)

    # List all JPEG files in the input directory
    jpeg_files = [f for f in os.listdir(input_folder) if f.endswith('.jpeg')]

    # Custom sort key function for JPEG files in the format fr15457985_ts833846284597_fe8.jpeg
    def sort_key(filename):
        parts = filename.split('_')
        fr_part = int(parts[0][2:])
        ts_part = int(parts[1][2:])
        fe_part = int(parts[2][2:].split('.')[0])
        return fe_part

    # Sort files using the custom sort key
    jpeg_files.sort(key=sort_key)
        # Calculate the middle index and select the middle 10,000 files
    middle_start = len(jpeg_files) // 2 - 5000
    middle_end = middle_start + 10000
    middle_jpeg_files = jpeg_files[middle_start:middle_end]

    # Copy the selected files to the output directory
    for file_name in middle_jpeg_files:
        source_path = os.path.join(input_folder, file_name)
        destination_path = os.path.join(output_folder, file_name)
        shutil.copy2(source_path, destination_path)

    print(f"Copied {len(middle_jpeg_files)} JPEG files to the output directory.")


# Example usage
input_folder = '/Zenodo Inconel Spiral/Image F/Im'
output_folder = '/Raw Samples Dataset/Zenodo Inconel Spiral'
copy_middle_jpegs(input_folder, output_folder)
```
